```python
df.date = pd.to_datetime(df.date)
df.dtypes
```

```python
# Adding a new column "year" in the dataframe that only contains year
df['year'] = df.date.dt.year
df
```

```python
# Adding another column that only contains the month
df['month'] = df.date.dt.month
df
```

```python
# Position the month column as the second one on the dataset
df.insert(1 , 'months' , df.date.dt.month)
df
```

```python
# Removing the 'year' and 'month' columns from the dataset
df.drop(['month', 'year'], axis=1, inplace = True)
```
